```r
q<-count(a,team1)
r1<-count(a,team2)
mu_data<-mutate(q,total=q$n+r1$n)
mu_data<-mu_data[,-2]
mu_data
```

```r
b<-count(a,winner)
r<-filter(b,b$winner!="")
arrange(r,desc(r$n))
r<-mutate(r,total=mu_data$total)
win.per<-mutate(r,win_per=r$n/mu_data$total)
arrange(win.per,desc(win.per$win_per))
```

```r
p<-select(a,toss_winner,toss_decision,winner)
k<-head(p)
f<-length(which(as.character(p$toss_winner)==as.character(p$winner)))
g<-length(p$toss_winner)
c<-c(f/g,1-f/g)
c
colors<-c("Green","Blue")
pie(c,init.angle = 90,label = c("Toss won","Toss lose"),col = colors)
legend("bottomright", c("Toss won percent","Toss lose percent"), fill = colors) 
```
